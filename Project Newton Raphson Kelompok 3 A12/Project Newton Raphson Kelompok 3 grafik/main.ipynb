### Import Library

```python
from newtonraphson_library import NewtonRaphson
import math
```

## Contoh 1 — Fungsi Polinomial

```python
f1 = lambda x: x**3 + x**2 - 3*x -3
# mendefinisikan fungsi f(x)
df1 = lambda x: 3*x**2 + 2*x - 3
# mendefinisikan turunan fungsi f'(x)

nr1 = NewtonRaphson(
    fungsi=f1,
    # memasukkan fungsi f(x) ke dalam class

    turunan=df1,
    # memasukkan turunan f'(x)

    x0=1.0,
    # menentukan tebakan awal (initial guess)

    toleransi=0.0001,
    # menentukan batas galat (error) agar iterasi berhenti

    maks_iter=50
    # menentukan jumlah maksimum iterasi
)

nr1.hitung()
# menjalankan proses iterasi metode Newton-Raphson

nr1.tampilkan_hasil()
# menampilkan tabel hasil iterasi (x lama, x baru, galat, dll)
nr1.analisis()
# menampilkan analisis hasil:
# apakah konvergen, jumlah iterasi, galat terakhir, dll

#Tampilkan Visualisasi Dashboard 
print("\nMenampilkan Dashboard Visualisasi (Subplots)...")
nr1.plot_visualisasi(rentang=3)
```

## Contoh 2 — Fungsi Trigonometri

```python
f2 = lambda x: math.cos(x) - x
# mendefinisikan fungsi f(x) 

df2 = lambda x: -math.sin(x) - 1
# mendefinisikan turunan f'(x)


nr2 = NewtonRaphson(
    fungsi=f2,
     # memasukkan fungsi f(x) ke dalam class

    turunan=df2,
    # memasukkan turunan f'(x)

    x0=0.5,
    # menentukan tebakan awal (initial guess)

    toleransi=0.001,
    # menentukan batas galat (error) agar iterasi berhenti

    maks_iter=50
    # menentukan jumlah maksimum iterasi
)

nr2.hitung()
# menjalankan iterasi metode Newton-Raphson
# mencari akar dari persamaan cos(x) - x = 0

nr2.tampilkan_hasil()
# menampilkan tabel hasil iterasi (nilai x, f(x), galat, dll)

nr2.analisis()
# menampilkan analisis hasil:
# apakah metode konvergen atau tidak
# serta informasi jumlah iterasi dan galat


#Tampilkan Visualisasi Dashboard 
print("\nMenampilkan Dashboard Visualisasi (Subplots)...")
nr2.plot_visualisasi(rentang=3)

```
